```python
import pandas as pd

# Load your newly created master file
flood_df = pd.read_csv('../data/processed/flood_master.csv')

print(f"Rows: {flood_df.shape[0]}")
print(f"Columns: {flood_df.shape[1]}")
print(f"Column names:\n{list(flood_df.columns)}")
print(f"\nFirst row:\n{flood_df.iloc[0]}")
print(f"\nMissing values:\n{flood_df.isnull().sum()}")
```

```python
print(flood_df['Area Affected'].unique())
print(flood_df['Severity'].unique())
print(flood_df['DFSI'].min(), flood_df['DFSI'].max())
print(flood_df['Human_fatality'].unique())
```

```python
# Create binary target for flood (severe/moderate) based on DFSI index
flood_df['target'] = (flood_df['DFSI'] >= 8.0).astype(int)
print("Class balance:", flood_df['target'].value_counts())
```
