```python
import geopandas as gpd
import folium

# Load road lines
roads = gpd.read_file(
    r'C:\SEM III\GBM_sih\data\raw\roads\hotosm_ind_roads_lines_geojson.geojson'
)

# Quick preview
print(roads.head())
print(roads.columns)

# Make a map for a region (e.g., Delhi NCR)
m = folium.Map(location=[28.6139, 77.2090], zoom_start=8)
folium.GeoJson(roads).add_to(m)
m.save('roads_preview.html')
```
